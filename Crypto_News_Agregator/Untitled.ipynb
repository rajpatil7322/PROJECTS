```python
from pycoingecko import CoinGeckoAPI
import pandas as pd



cg = CoinGeckoAPI()


#print(cg.get_price(ids='bitcoin', vs_currencies='usd'))
volume=cg.get_exchanges_volume_chart_by_id(id='binance',days=100)

vol=pd.DataFrame(data=volume,columns=['target','volume'])
print(vol.head())
```

```python
import matplotlib.pyplot as plt
plt.plot(vol['target'])
```
